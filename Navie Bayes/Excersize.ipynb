```python
import numpy as np
import pandas as pd
import sklearn
```

```python
mr = pd.read_csv('movie_review_train.csv')
mr.head()
```

```python
mr['class'].value_counts()
```

```python
mr['class'] = mr['class'].map({'Pos':1,'Neg':0})
mr.head
```

```python
train_array = mr.values
train_array
```

```python
X_train = train_array[:,1]
y_train = train_array[:,0]
print(X_train)
print(y_train)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vec = CountVectorizer(stop_words='english')
vec.fit(X_train)
vec.vocabulary_
```

```python
print(len(vec.get_feature_names()))
```

```python
dict2 = CountVectorizer(stop_words='english', min_df=.03, max_df=.8)
dict2.fit(X_train)
print(len(dict2.get_feature_names()))
```

```python
X_trns1 = dict2.transform(X_train)
print(X_trns1)
```

```python
test_doc = pd.read_csv('movie_review_test.csv')
test_doc.head()
```

```python
test_doc['class']= test_doc['class'].map({'Pos':1,'Neg':0})
test_doc.head()
```

```python
test_values = test_doc.values
X_test = test_values[:,1]
y_test = test_values[:,0]
```

```python
X_test_trans = dict2.transform(X_test)
X_test_trans
```

```python
X_test1 = X_test_trans.toarray()
X_test1
```

```python
from sklearn.naive_bayes import BernoulliNB

bnb = BernoulliNB()

prob_bnb = bnb.predict_proba(X_test1)
prob_bnb
```
